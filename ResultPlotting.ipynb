```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
sns.set()
```

```python
#Learning
HopfLearning = np.array(pd.read_csv('HopfLearning.csv'))
KuramotoLearning = np.array(pd.read_csv('KuramotoLearning.csv'))
XYLearning = np.array(pd.read_csv('XYLearning.csv'))
GCNLearning = np.array(pd.read_csv('GCNLearning.csv'))
```

```python
df = pd.DataFrame({'GCN': GCNLearning[:,1], 'Kuramoto': KuramotoLearning[:,1],'Hopf': HopfLearning[:,1], 'XY': XYLearning[:,1]})
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=False)
g = sns.lineplot(ax=axes[0],data=df)
g.set_xlabel('Epoch')
g.set_ylabel('Validation Accuracy')
df = pd.DataFrame({'GCN': GCNLearning[:,2], 'Kuramoto': KuramotoLearning[:,2],'Hopf': HopfLearning[:,2], 'XY': XYLearning[:,2]})
g = sns.lineplot(ax=axes[1],data=df)
g.set_xlabel('Epoch')
g.set_ylabel('Test Accuracy')
df = pd.DataFrame({'GCN': GCNLearning[:,0], 'Kuramoto': KuramotoLearning[:,0],'Hopf': HopfLearning[:,0], 'XY': XYLearning[:,0]})
g = sns.lineplot(ax=axes[2],data=df)
g.set_xlabel('Epoch')
g.set_ylabel('Training Loss')
```
